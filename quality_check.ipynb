```python
from matplotlib import pyplot as plt
import os
import pandas as pd

from pynwb import NWBHDF5IO
```

```python
dir_path = '000252'
neuron_list = []
file_list = []
for root, dirs, files in os.walk(dir_path):
    for file in files:
        if file.endswith('.nwb'):
            file_list.append(file)
            nwb_file = os.path.join(root, file)
            
            with NWBHDF5IO(nwb_file, 'r') as io:
                nwb = io.read()
                neuron_list.append(nwb.units.to_dataframe().shape[0])
                io.close()
```

```python
stats_df = pd.DataFrame()

stats_df['file'] = file_list
stats_df['# neurons'] = neuron_list

stats_df
```
